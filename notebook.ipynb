```python
fundamental_data = fundamental_data.dropna()
```

```python
fundamental_data.isnull().sum()
```

```python
# visualize market data
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(market_data['Close'], label='Close Price')
plt.title('Market Data - Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
# Merge datasets on a common column, e.g., 'ticker'
merged_data = pd.merge(market_data, fundamental_data, on = 'Ticker', how = 'inner')
```

```python
#save the merged data to a new CSV file
merged_data.to_csv("merged_stock_data.csv", index=False)
```
